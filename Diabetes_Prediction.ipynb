```python
def iqr_method(df,variables):
  q1=df[variables].quantile(0.25)
  q3=df[variables].quantile(0.75)
  iqr=q3-q1
  upper=q3+(1.5*iqr)
  lower=q1-(1.5*iqr)
  return lower,upper
```

```python
lst=['age', 'hypertension', 'heart_disease', 'smoking_history', 'bmi',
       'HbA1c_level', 'blood_glucose_level', 'diabetes']
```

```python
for i in lst:
  print(i)
  lower_lim,upper_lim=iqr_method(df,i)
  print('lower limit = ',lower_lim)
  print('upper limitÂ =Â ',upper_lim)
  df[i]=np.where(df[i]>upper_lim,upper_lim,
                   np.where(df[i]<lower_lim,lower_lim,df[i]))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree  import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

knn=KNeighborsClassifier(n_neighbors=7)
tree=DecisionTreeClassifier(criterion='entropy')
ada=AdaBoostClassifier(n_estimators=50,random_state=42)
clas=RandomForestClassifier(n_estimators=10,criterion='entropy')
lst=[knn,tree,ada,clas]
```

```python
for i in lst:
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print("Model is ",i)
  print("y_pred is",y_pred)
  print("*"*100)
  print('confusion_matrix is',confusion_matrix(y_test,y_pred))
  print("*"*100)
  print("Accuracy score is",accuracy_score(y_test,y_pred))
  print("*"*100)
  print('classifictaion report is',classification_report(y_test,y_pred))
```

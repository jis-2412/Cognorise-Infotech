```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['street']=le.fit_transform(df['street'])
df['statezip']=le.fit_transform(df['statezip'])
df['country']=le.fit_transform(df['country'])
df['city']=le.fit_transform(df['city'])
df['date']=le.fit_transform(df['date'])
df
```

```python
df.drop(['country','street','statezip','date'],axis=1,inplace=True)
```

```python
x=df.iloc[:,1:].values
x
```

```python
y=df.iloc[:,0].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import  mean_squared_error, r2_score

lin=LinearRegression()
tree = DecisionTreeRegressor(criterion='squared_error',random_state=1)
ran= RandomForestRegressor(random_state=1)
gra = GradientBoostingRegressor()
```

```python
lst=[lin,tree,ran,gra]
for i in lst:
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print("Model is ",i)
  print("y_pred is",y_pred)
  print("*"*100)
  print('Mean Squared Error is', mean_squared_error(y_test, y_pred))  # Use MSE
  print("*" * 100)
  print("R-squared score is", r2_score(y_test, y_pred))  # Use R-squared
  print("*" * 100)
```
